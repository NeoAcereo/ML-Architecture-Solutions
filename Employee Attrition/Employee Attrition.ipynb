## Data Preparation and Preprocessing

```python
age = dataset['Age']

age_ranges = pd.IntervalIndex.from_tuples([(18, 30), (31, 40), (41, 50), (51, 60)], closed='both')

labels = ['Kid', 'Youth', 'Adult', 'Elderly']
age_groups = pd.cut(x=age, bins=age_ranges)
age_groups = age_groups.cat.rename_categories(labels)
```

```python
dataset['AgeGroup'] = age_groups
dataset = dataset.drop('Age', axis=1)
dataset['AgeGroup']
```

```python
from category_encoders import OrdinalEncoder

mapping = [{   
  "col": "AgeGroup",
    "mapping": {
        'Kid': 1,
        'Youth': 2,
        'Adult': 3,
        'Elderly':4
    }
    }
]
ordinal_encoder = OrdinalEncoder(cols='AgeGroup', mapping=mapping)
age_group = ordinal_encoder.fit_transform(dataset['AgeGroup'])
dataset['AgeGroup'] = age_group
dataset['AgeGroup']
```

## Model Training and Evaluation

```python
from sklearn.model_selection import train_test_split
feature_train, feature_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
feature_train
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

svc = SVC(degree=4)
lr_classifier = LogisticRegression(max_iter=210)


tree_classifier = DecisionTreeClassifier(random_state=42, ccp_alpha=0.0045)
```

```python
svc.fit(feature_train, target_train)
svc.score(feature_test, target_test)
```

```python
lr_classifier.fit(feature_train, target_train)
lr_classifier.score(feature_test, target_test)
```

```python
tree_classifier.fit(feature_train, target_train)
tree_classifier.score(feature_test, target_test)
```

```python
class AutoVivification(dict): # dynamic dictionary
    def __getitem__(self, item):
        try:
            return dict.__getitem__(self, item)
        except KeyError:
            value = self[item] = type(self)()
            return value
```

```python
model_scores = AutoVivification()
def evaluate_models(models, scorers, X_test, y_test):
    """Evaluate trained models on given scores and prints out the metric score of each trained model.
    models:
        Must be a list of tuples with each tuple having the model name and the trained model instance.
    scorers:
        Must be a list of tuples with each tuple having the score name and the scorer instance.
    X_test:
        The feature test set. Must be of size (n_features, n_samples)
    y_test:
        The target test set. Must be of size (n_samples,)
     """
    for(model_name, model) in models:
        print(f'Test Scores of {model_name} ')
        for (score_name, scorer) in scorers:        
            test_score = scorer(y_true=y_test, y_pred=model.predict(X_test))
            model_scores[model_name][score_name] = test_score
            print(f'{score_name} : {test_score:.4f}')
        print('\n')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

scorers = [('accuracy', accuracy_score), ('precision', precision_score), ('recall', recall_score)]
models = [('Logistic Regression', lr_classifier), ('SVM', svc), ('Decision Tree', tree_classifier)]

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)
```

## Using Ensembles

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, VotingClassifier, BaggingClassifier, StackingClassifier
```

### AdaBoost

```python
lr_adaboost = AdaBoostClassifier(estimator=lr_classifier,
                                      n_estimators=100)

tree_adaboost = AdaBoostClassifier(estimator=tree_classifier,
                                   n_estimators=100)

lr_adaboost.fit(feature_train, target_train)
tree_adaboost.fit(feature_train, target_train)

models = [('Logistic Regression AdaBoost', lr_adaboost), ('Decision Tree AdaBoost', tree_adaboost)]

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)    
```

### Blender 

```python
estimators = [('Logistic Regression AdaBoost', lr_adaboost), ('Decision Tree AdaBoost', tree_adaboost)]

blender = VotingClassifier(estimators=estimators, n_jobs=-2)
blender.fit(feature_train, target_train)

models = [('Blender', blender)]

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)
```

### Bagging Classifier

```python
rf_classifier = RandomForestClassifier(n_estimators=100, max_features='sqrt') # bagging for decision trees

rf_classifier.fit(feature_train, target_train)

models = []

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)
lr_bagging = BaggingClassifier(estimator=lr_classifier,
                                   n_estimators=100,
                                   random_state=42,
                                   bootstrap=True,
                                   bootstrap_features=True,
                                   n_jobs=-2)

svm_bagging = BaggingClassifier(estimator=svc,
                                   n_estimators=100,
                                   random_state=42,
                                   bootstrap=True,
                                   bootstrap_features=True,
                                   n_jobs=-2)

lr_bagging.fit(feature_train, target_train)
svm_bagging.fit(feature_train, target_train)

models = [('Random Forest', rf_classifier), ('Logistic Regression Bagging', lr_bagging), ('SVM Bagging', svm_bagging)]

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)
```

### Stacking Classifier

```python
svm_estimators=[('Logistic Regression ',lr_classifier), ('Decision Tree ', tree_classifier)]

svm_stacker = StackingClassifier(estimators=svm_estimators, final_estimator=svc)
svm_stacker.fit(feature_train, target_train)

lr_estimators = [('SVM', svc), ('Decision Tree', tree_classifier)]
lr_stacker = StackingClassifier(estimators=lr_estimators, final_estimator=lr_classifier)
lr_stacker.fit(feature_train, target_train)

tree_estimators = [('Logisitic Regression', lr_classifier), ('SVM', svc)]
tree_stacker = StackingClassifier(estimators=tree_estimators, final_estimator=tree_classifier)
tree_stacker.fit(feature_train, target_train)

models = [('SVM Stacker', svm_stacker), ('Logistic Regression Stacker', lr_stacker),
          ('Decision Tree Stacker', tree_stacker)]

evaluate_models(models=models, scorers=scorers, X_test=feature_test, y_test=target_test)
```

```python
estimators=[('RandomForest',rf_classifier), ('Decision Tree AdaBoost', tree_adaboost)]

stacker = StackingClassifier(estimators=estimators, final_estimator=lr_adaboost)
stacker.fit(feature_train, target_train)

models = [('Logistic Regression AdaBoost Stacker', stacker)]

for (score_name, scorer) in scorers:
    for(model_name, model) in models:
        test_score = scorer(y_true=target_test, y_pred=model.predict(feature_test))
        model_scores[model_name][score_name] = test_score
        print(f'The test {score_name} of {model_name} is {test_score:.4f}')
    print('*'*10)
```
