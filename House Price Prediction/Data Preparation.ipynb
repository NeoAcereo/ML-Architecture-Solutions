```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = dataset.drop(['Id'], axis=1)
```

```python
dataset_2 = dataset
```

```python
temp = dataset.isnull().any().astype(str)
```

```python
na_columns = []
for column in dataset.columns:
    if(temp[column] == 'True'):
        na_columns.append(column)
```

```python
na_columns
```

```python
dataset.astype(str)
```

```python
# taking out all features with 80% or
# more of their values missing

for column in na_columns:
    na_count=0
    for value in dataset[column].astype(str): 
    
        if (value == 'nan'):
            na_count +=1
            
    print(f'{column}: nan_count:{na_count}')        
    if(na_count >= 0.8 * len(dataset)):
        dataset_2 = dataset_2.drop([column], axis=1)
```

```python
dataset_2
```

```python
dataset_2['LotFrontage'].head(50)
```

```python
#filling missing values with mean
dataset_2['LotFrontage'] = dataset_2['LotFrontage'].fillna(dataset_2['LotFrontage'].mean())
```

```python
dataset_2['LotFrontage'].head(50)
```

```python
temp_2 = dataset_2.isna().any().astype(str)

na_columns_2 = []


for column in dataset_2.columns:
    if(temp_2[column] == 'True'):
        na_columns_2.append(column)
```

```python
# filling missing values with mode
for column in na_columns_2:
    dataset_2[column] = dataset_2[column].fillna(dataset_2[column].mode())
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
def one_hot_encode(data,feature,keep_first=True):

    oh = OneHotEncoder()
   
    oh_df = pd.DataFrame(oh.fit_transform(data[[feature]]).toarray()) 
    
    oh_df.columns = oh.get_feature_names()
    
    for col in oh_df.columns:
        oh_df.rename({col:f'{feature}_'+col.split('_')[1]},axis=1,inplace=True)
    
    new_data = pd.concat([data,oh_df],axis=1)
    new_data.drop(feature,axis=1,inplace=True)
    
    if keep_first == False:
        new_data=new_data.iloc[:,1:]
    
    return new_data
```

```python
def scale_feature(data,feature):
    scaler = StandardScaler()
    scaled_df = pd.DataFrame(scaler.fit_transform(data[[feature]]), columns=[f'scaled_{feature}'])
    
    new_data = pd.concat([data,scaled_df], axis=1).drop([feature], axis=1)
    return new_data
    
```

```python
continuous_features = dataset_2.select_dtypes(include=['int', 'float'])
```

```python
categorical_features = dataset_2.select_dtypes(include='object')
```

```python
for column in categorical_features.columns:
    categorical_features = one_hot_encode(categorical_features, column)
```

```python
categorical_features
```

```python
for column in continuous_features.columns:
    continuous_features = scale_feature(continuous_features, column)
```

```python
continuous_features
```

```python
new_dataset = pd.concat([continuous_features, categorical_features], axis=1)
```
