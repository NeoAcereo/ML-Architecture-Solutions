```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
```

```python
X, y = fetch_openml(
    "titanic", version=1, as_frame=True, return_X_y=True
)
```

```python
X
```

```python
import numpy as np

random_number_generator = np.random.RandomState(seed=42)
```

```python
X["random_cat"] = random_number_generator.randint(3, size=X.shape[0])
X["random_num"] = random_number_generator.randn(X.shape[0])
```

```python
categorical_columns = ["pclass", "sex", "embarked", "random_cat"]
numerical_columns = ["age", "sibsp", "parch", "fare", "random_num"]
```

```python
X
```

```python
X = X[categorical_columns + numerical_columns]
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
X_train
```

```python
import seaborn as sns
sns.countplot(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
```

```python
categorical_encoder = OrdinalEncoder(
    handle_unknown="use_encoded_value", unknown_value=-1
)

numerical_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())

])

categorical_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal_encoder', OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1))
])

preprocessor = ColumnTransformer([
    ('cat', categorical_pipe, categorical_columns),
    ('num', numerical_pipe, numerical_columns)
])


rf = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42)),
    
])

rf.fit(X_train, y_train)
```

```python
X_train
```

```python
y_train
```

```python
print(f'Rf train score: {rf.score(X_train, y_train):0.3f}')
print(f'Rf test score: {rf.score(X_test, y_test):0.3f}')
```

```python
import pandas as pd
feature_names = rf[:-1].get_feature_names_out()
feature_importances = rf[-1].feature_importances_
```

```python
len(feature_names)
```

```python
mean_decrease_impurity = pd.Series(
    data=feature_importances, index=feature_names).sort_values(ascending=True)
```

```python
ax = mean_decrease_impurity.plot.barh()
ax.set_title('Random Forest MDI Feature Importance')
ax.figure.tight_layout()
```

```python
rf.set_params(classifier__min_samples_leaf=20).fit(X_train, y_train) # decrease examples for each leaf to reduce overfitting
```

```python
print(f"RF train accuracy: {rf.score(X_train, y_train):.3f}")
print(f"RF test accuracy: {rf.score(X_test, y_test):.3f}")
```

```python
train_importances = permutation_importance(rf, X_train, y_train, n_repeats=10, n_jobs=-1, random_state=42)
test_importances = permutation_importance(rf, X_test, y_test, n_repeats=10, n_jobs=-1, random_state=42)
```

```python
sorted_train_index = train_importances.importances_mean.argsort()
sorted_test_index = test_importances.importances_mean.argsort()
```

```python
train_results = pd.DataFrame(data=train_importances.importances[sorted_train_index].T, columns=X.columns[sorted_index])
test_results = pd.DataFrame(data=test_importances.importances[sorted_test_index].T, columns=X.columns[sorted_index])
```

```python
train_results
```

```python
test_results
```

```python
for name, importance in zip(['train', 'test'], [train_results, test_results]):
    ax = importance.plot.box(vert=False, whis=10)
    ax.set_title(f'Permutation importance on {name} set')
    ax.axvline(x=0, color='red', linestyle=':')
    ax.set_xlabel(f'Decrease in {name} accuracy')
    ax.figure.tight_layout()
```
