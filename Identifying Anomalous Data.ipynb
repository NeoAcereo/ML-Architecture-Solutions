## Using a univariate model

```python
from sklearn.datasets import load_boston
import pandas as pd
%matplotlib inline 
boston = load_boston()
X, y = boston.data, boston.target
X = pd.DataFrame(X, columns=boston.feature_names)
X.boxplot('LSTAT',return_type='axes')
```

## Using a multivariate model

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca_features = PCA().fit_transform(StandardScaler().fit_transform(X))
```

```python
import matplotlib.pyplot as plt
#most information from the dataset is concentrated on the first two components
plt.scatter(pca_features[:,0],pca_features[:,1], s=2**7, edgecolors='white',
            alpha=0.85, cmap='autumn')
plt.grid() 
plt.xlabel('Component 1') 
plt.ylabel('Component 2') 
```

```python
# anomalous data is more concentrated in the last layers of the pca values
plt.scatter(pca_features[:,11], pca_features[:,12], s=2**7, edgecolors='red', alpha=0.85, cmap='autumn')
plt.grid()
plt.xlabel('Component 11')
plt.ylabel('Component 12')
```
