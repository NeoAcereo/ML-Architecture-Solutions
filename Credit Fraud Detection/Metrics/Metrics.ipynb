```python
features = dataset.drop(['Class'], axis=1)
target = dataset['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=0) 
X_test
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python

numerical_columns = dataset.drop('Class', axis=1).columns

numerical_pipe = Pipeline([
    ('imputing', SimpleImputer(strategy='mean')),
    
    ('scaling', StandardScaler())
])

preprocessing = ColumnTransformer([
    ('numerical transformation', numerical_pipe, numerical_columns)
])

rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)

ada_classifier = AdaBoostClassifier(estimator=LogisticRegression(), n_estimators=300, random_state=42)

models = [rf_classifier, ada_classifier]

model_labels = ['RandomForest', 'AdaBoost']

model_pipes = dict()

for model_label, model in zip(model_labels, models):
    model_pipes[model_label] =  Pipeline([
    ('preprocessing', preprocessing),
    ('classification', model)
])


```

```python
import sklearn.metrics as metrics

for model in model_labels:
    model_pipes[model].fit(X_train, y_train)
    predictions = model_pipes[model].predict(X_test)
    score = metrics.f1_score(y_test, predictions)
    print(f'f1-score for {model} is {score: .4f}')
```
