```python
from sklearn.utils import resample

majority_dataset = dataset[dataset['Class'] == 0] 
minority_dataset = dataset[dataset['Class'] == 1] 

n_samples = round(len(majority_dataset)/2)

undersampled_majority_dataset = resample(majority_dataset, 
                                 replace=False,    
                                 n_samples= n_samples,
                                 random_state=42)


oversampled_minority_dataset = resample(minority_dataset, 
                                 replace=True,    
                                 n_samples= n_samples, 
                                 random_state=42)  

resampled_dataset = pd.concat([ oversampled_minority_dataset, undersampled_majority_dataset])

```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python

numerical_columns = resampled_dataset.drop('Class', axis=1).columns

numerical_pipe = Pipeline([
    ('imputing', SimpleImputer(strategy='mean')),
    
    ('scaling', StandardScaler())
])

preprocessing = ColumnTransformer([
    ('numerical transformation', numerical_pipe, numerical_columns)
])

rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)

ada_classifier = AdaBoostClassifier(estimator=LogisticRegression(), n_estimators=300, random_state=42)

models = [rf_classifier, ada_classifier]

model_labels = ['RandomForest', 'AdaBoost']

model_pipes = dict()

for model_label, model in zip(model_labels, models):
    model_pipes[model_label] =  Pipeline([
    ('preprocessing', preprocessing),
    ('classification', model)
])


```

```python
class AutoVivification(dict):
    """Implementation of perl's autovivification feature."""
    def __getitem__(self, item):
        try:
            return dict.__getitem__(self, item)
        except KeyError:
            value = self[item] = type(self)()
            return value
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix



def calculate_classification_metrics(X_train, y_train, X_test, y_true, metrics, models):
    
    """
    Calculates performance metrics of chosen classifiers. 
    
    Parameters:
    
        X_train:
        The features values used for training the algorithm.
        
        y_train:
        The target feature values used for training the algorithm.
        
        X_test:
        The values of features used for the test predictions.
    
        y_true:
        The expected values of the target feature.
        
        
        metrics:
        a dictionary having the names of the metrics as keys and the metric objects as values.
        
        models:
        a dictionary having the names of the models as keys and the model objects or instances as values.
        
    returns:
        A nested dictionary of the models and their scores. The model names serve as the keys for the
        first level with the metric names serving as the keys at the second level.
    """
    model_scores = AutoVivification()
    
    for model in models:
        models[model].fit(X_train, y_train)
        for metric in metrics:
            predictions = models[model].predict(X_test)
            score = metrics[metric](y_true, predictions)
            model_scores[model][metric] = score
            
    return model_scores
```

```python

metrics = {'accuracy':accuracy_score, 'precision':precision_score, 'recall':recall_score}

model_scores = calculate_classification_metrics(X_train= X_train,
                                                y_train= y_train,
                                                X_test=X_test,
                                                y_true=y_test, 
                                                metrics=metrics,
                                                models=model_pipes)
```
